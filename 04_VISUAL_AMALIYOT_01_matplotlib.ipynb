```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20,10))
plt.plot(df.index, df[["T"]], color="b", marker="h", markerfacecolor="r", linestyle="--")
plt.grid(color="#C4DFAA", linestyle="--")
plt.ylabel("average T")
plt.xlabel("Year")
plt.legend(["average T"], loc=2, fontsize="x-large")
plt.title("Average Annual Temperature in Uzbekistan")

plt.show()
```

## 8-VAZIFA. Toshkent shahri uchun, har bir yilda o'rtacha min va max haroratlarni ko'rsatuvchi grafik chizing. **Grafikni faylga saqlab oling**

```python
# JAVOBNI SHUYERGA YOZING
plt.figure(figsize=(20, 10))

plt.plot(df.index, df["TM"],
         linewidth=1.5,
         linestyle="--",
         color="#FF9F29",
         alpha=1,
         marker="o")

plt.plot(df.index, df["Tm"],
         linewidth=2,
         linestyle="--",
         color="#18978F",
         alpha=0.7,
         marker="h")
plt.title("Max va Min T in Tashkent")
plt.ylabel('Temperature')
plt.xlabel("Years")
plt.legend(['Max', 'Min'])
plt.grid(color="#3AB0FF", linestyle="--")
plt.savefig("graph Temperatur.png")
plt.show()
```

```python
#@title To'g'ri javob
minT = df['Tm']
maxT = df['TM']
plt.figure(figsize=(16,8))
plt.plot(minT, linewidth=2, linestyle="--", marker='h', markerfacecolor='g', label="O'rtacha minimum harorat")
plt.plot(maxT, linewidth=2, linestyle="-", marker='v', markerfacecolor='r', label="O'rtacha maksimum harorat")
plt.grid()
plt.legend()
plt.title("Toshkent shahri uchun yillik o'rtacha min va max harorat")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.savefig('tempMinMax.png')
plt.show()
```
